```python
k = 10
value_user_matrix = np.zeros((k,cf3.similarity_matrix.shape[1]))
index_user_matrix = np.zeros((k,cf3.similarity_matrix.shape[1]))
for i in range(cf3.similarity_matrix.shape[1]):
    temp = cf3.similarity_matrix[:,i]
    ind = np.argpartition(temp, -k)[-k:]
    index_user_matrix[:,i] = ind
    value_user_matrix[:,i] = temp[ind]
```

```python
index_user_matrix
```

```python
value_user_matrix
print(value_user_matrix.shape)
```

```python
#k_similar_users_profiles = np.zeros((k,cf3.utility_matrix.shape[1]))
#k_similar_users_similarity_values = np.zeros(k)
final_user_item_score = np.zeros(cf3.utility_matrix.shape)
for i in range(index_user_matrix.shape[1]):
    temp = index_user_matrix[:,i].astype(int)
    k_similar_users_profiles = cf3.utility_matrix[temp]
    k_similar_users_similarity_values = value_user_matrix[:,i]
    final_user_item_score[i,:] = np.dot(k_similar_users_similarity_values,k_similar_users_profiles)
```
